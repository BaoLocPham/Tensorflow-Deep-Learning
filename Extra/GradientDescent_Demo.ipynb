```python
from matplotlib import pyplot as plt 
import numpy as np 
from matplotlib.animation import FuncAnimation 
```

```python
def plot_function():
  x = np.linspace(-5, 5, 100)

  y = cost(x)

  fig = plt.gcf()
  fig.suptitle("f(x)=x^2 + 5 sin(x)", fontsize=14)
  plt.plot(x, y,'grey')
```

```python
def grad(x):
  return 2*x + 5*np.cos(x)

def cost(x):
  return x**2 + 5*np.sin(x)
```

```python
x1 = -4
y1 = cost(x1)
x2 = 2
y2 = cost(x2)

plot_function()
plt.scatter(x1,y1,c='k')
plt.scatter(x2,y2,c='r')
plt.plot((x1,x2),(y1,y2), c='k')
plt.xlabel("con cac")
plt.show()
```

```python
def GD(alpha, x0):
  x = [x0]
  for i in range(100):
    x_new = x[-1] - alpha*grad(x[-1])
    if abs(grad(x_new))<1e-3:
      break
    x.append(x_new)
  return (x, i)
```

```python
(x1, circle1) = GD(.1,-5)
(x2, circle2) = GD(.1,5)

print("x1={:.2f}, cost={:.2f}, obtained after {} iterations".format(x1[-1], cost(x1[-1]), circle1))
print("x1={:.2f}, cost={:.2f}, obtained after {} iterations".format(x2[-1], cost(x2[-1]), circle2))
```

```python
def GD2(alpha, x0):
  x = [x0]
  y = [cost(x0)]
  for i in range(100):
    x_new = x[-1] - alpha*grad(x[-1])
    if abs(grad(x_new))<1e-3:
      break
    x.append(x_new)
    y.append(cost(x_new))
  return (x, y, i)
```
